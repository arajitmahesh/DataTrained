# Outliers

Let us now check for outliers in our dataset

```python
columns=['Survived','Pclass','Age','SibSp','Parch','Fare']
for i in columns:
    sns.boxplot(df[i])
    plt.show()
```

We can observe from the above boxplots that there are outliers in the features Age, SibSp, Parch and Fare.\
But we will not remove the outliers only from SibSp and Parch because it is discrete data.\
The features Age and Fare are continuous data, so let us remove the outliers from these features.

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df['Cabin'] = df['Cabin'].fillna('U')
```

df.isnull().sum()

# Skewness

Let us now check if there is any skew in the data

```python
new_df.skew()
```

We can see that the features Pclass, SibSp, Parch and Fare have skew data

We do not remove skewness from Pclass because it is categorical data.

Let us use the yeo-johnson method to remove the skewed data

```python
from sklearn.preprocessing import PowerTransformer
skewed_features = ["SibSp","Parch","Fare"]
scaler = PowerTransformer(method='yeo-johnson')
```

```python
new_df[skewed_features] = scaler.fit_transform(new_df[skewed_features].values)
new_df[skewed_features].head()
```

Let us now plot the distribution plots to see how the data is distributed

# Scale data using MinMax Scalar

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x = pd.DataFrame(mms.fit_transform(x), columns=x.columns)
```

```python
x.head()
```

# Finding best Random State

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state =i)
    DTC = DecisionTreeClassifier()
    DTC.fit(x_train, y_train)
    pred = DTC.predict(x_test)
    acc=accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu," on Random_state ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=maxRS)
```

```python
x.shape, y.shape
```

```python
x_train.shape,y_train.shape, x_test.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score, accuracy_score
```

# DecisionTreeClassifier

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train,y_train)
predDTC = DTC.predict(x_test)

print(accuracy_score(y_test, predDTC))
print(confusion_matrix(y_test, predDTC))
print(classification_report(y_test,predDTC))
```

We are getting an Accuracy of 84% using DecisionTreeClassifier

```python
cm = confusion_matrix(y_test,predDTC)

x_axis_labels = ["Died","Survived"]
y_axis_labels = ["Died","Survived"]

f , ax = plt.subplots(figsize=(7,7))
sns.heatmap(cm, annot = True,linewidths=.2, linecolor="black", fmt = ".0f", ax=ax, cmap="Purples", 
xticklabels=x_axis_labels,
yticklabels=y_axis_labels)

plt.xlabel("PREDICTED LABEL")
plt.ylabel("TRUE LABEL")
plt.title('Confusion Matrix for Decision Tree Classifier')
```

From the plot we can easily observe the predicted values and true values.

True-Positive = 140\
